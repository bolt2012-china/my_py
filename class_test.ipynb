```python
class SJTUers:
    def __init__(self): #__init__函数不是必须的，
                        # 如果没有，则说明这个类没有共同属性
        pass
    def eat(self):
        print("eating")
    def sleep(self):
        print("sleeping")
    def juan(self):
        print("juaning")
    def daily(self):
        self.eat()
        self.sleep()
        self.juan()         #或许这些函数才是共同属性也说不定

p = SJTUers()
print("SJTUers' daily",p.daily())
```

Triangle类

```python
class Triangle:
    def __init__(self, a, b, c): 
        self.a = a
        self.b = b
        self.c = c
        if self.a + self.b > self.c and self.b < self.a + self.c and self.b + self.c > self.a:
            print("you are a triangle")
        else:
            print("you aren't a triangle")
    
    def length(self):
        return self.a + self.b + self.c
    
    def area(self):
        p = (self.a + self.b + self.c)/2
        return (p*(p-self.a) * (p-self.b) * (p-self.c))**0.5
    
t = Triangle(2, 3, 6)
print(t.area())
```

Vector类

```python
import math
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.vec = (self.x, self.y)
        
    def add(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def dot_product(self, other):
        return self.x * other.x + self.y * other.y
    
    def length(self):
        return (self.x**2 + self.y**2)**0.5
    
    def angle(self, other):
        return math.atan2(self.y - other.y, self.x - other.x)
    
    def distance(self, other):
        return math.hypot(self.x - other.x, self.y - other.y)
    
    def is_perpendicular(self, other):
        if self.dot_product(other) == 0: return True
        return False
    
    def is_parallel(self, other):
        if self.angle(other) == 0: return True
        return False
v1 = Vector(2, 3)
v2 = Vector(4, 5)
print(Vector.is_perpendicular(v1,v2))
```

__单例__

new是一个内置函数，为新构造的类开辟空间。
它的返回值是构造的class的第一个实例化对象，即self.
未定义new时自动调用父类的new.

```python
class MusicPlayer1(object):
    #记录第一个被创建的对象的引用
    instance =None
    def __new__(cls, *args, **kwargs):
        #1.创建类属性是否为空对象
        if cls.instance is None:
            #2.调用父类的方法，为第一个对象分配空间
            cls.instance =super().__new__(cls)  #super用法其实就是调用父类的
            #也可以写作 return object.__new__(cls,*args,**kwargs)
        return cls.instance
        
        
class MusicPlayer2(object):
    #记录第一个被创建的对象的引用
    instance =None
    def __new__(cls, *args, **kwargs):
        #1.创建类属性是否为空对象
        if cls.instance is None:
            #2.调用父类的方法，为第一个对象分配空间
            cls.instance =super().__new__(cls)
        return cls.instance
    
play1 =MusicPlayer1()       #不同的类new分配的地址才能不同   
print(play1)
play3=MusicPlayer2()        #同一个类new分配的地址是一样的
print(play3)
play2 =MusicPlayer2()
print(play2)
```

__ALL ARE OBJECTS__——函数作为对象

```python
def greeting(name):
    hello = 'Hello, ' + name + '!'
    return hello

print(greeting('John'))
# Hello, John!
say_hello = greeting

print(say_hello('Mike'))
# Hello, Mike!
print(id(greeting))
# 1643325111520
print(id(say_hello))
# 1643325111520
# say_hello与greeting是对同一对象的不同引用
del greeting

# print(id(greeting)) #NameError: name 'greeting' is not defined
print(id(say_hello))
# 1643325111520
# 这些地址都是一样的，只有greeting的引用被解除了所以报错，而say_hello仍指向原来统一的地址
say_hello('Jennifer')
# Hello, Jennifer



```

函数的多种用法（参数，返回值，组成其他对象）

```python
#函数作为传参
def combine_two_numbers(how_to, numbers):
    return how_to(numbers)

def add_two_numbers(numbers):
    a, b = numbers
    return a + b

def multiply_two_numbers(numbers):
    a, b = numbers
    return a * b

combine_two_numbers(add_two_numbers, (3, 5))
# 8

combine_two_numbers(multiply_two_numbers, (4, 6))
# 24


#函数作为返回值
def add_number_creator(number):
    def add_number(a):
        return a + number
    return add_number

add_three = add_number_creator(3)
add_five = add_number_creator(5)

add_three(5)
# 8



#函数作为其他对象的组成，这里是list
add_functions = [add_number_creator(0), add_number_creator(1), add_number_creator(2)]

for i, func in enumerate(add_functions):
    a = 8
    print('a is 8, adding '+ str(i) + ' is' + str(func(a)))
    
# a is 8, adding 0 is 8
# a is 8, adding 1 is 9
# a is 8, adding 2 is 10
```

代码的多态：同一个函数，根据传入的参数不同返回不同的结果
